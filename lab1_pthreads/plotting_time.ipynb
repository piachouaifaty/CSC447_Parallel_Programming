```python
import matplotlib.pyplot as plt

plt.xlabel('array size')
plt.ylabel('runtime (ms)')

#serial
plt.plot([1000, 10000, 100000, 1000000, 10000000, 100000000], [0, 1, 1, 5, 44, 383], label="sequential")

#pthreads
plt.plot([1000, 10000, 100000, 1000000, 10000000, 100000000], [0, 0, 0, 2, 19, 197], label="pthreads")

plt.title('runtime as a function of array size for sequential and parallelized code')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```
